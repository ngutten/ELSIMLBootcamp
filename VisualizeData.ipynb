```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

# Dataset loader for IRIS
from sklearn.datasets import load_iris
```

```python
iris_data = load_iris()

x = iris_data['data']
y = iris_data['target']
features = iris_data['feature_names']

print("Dataset has %d rows, %d features, %d unique labels" % (x.shape[0], x.shape[1], np.unique(y).shape[0]))
```

```python
# If we have a small number of features, we can get a sense of the dataset 
# by plotting each feature against each other.
#
# Can't really do this if we have more than 10 or so...

for i in range(4):
    for j in range(4):
        plt.subplot(4,4,1+i+4*j)
        plt.scatter(x[:,i], x[:,j], c=y, cmap=plt.cm.Set1)
        if j==3:
            plt.xlabel(features[i])
        if i==0:
            plt.ylabel(features[j])

plt.gcf().set_size_inches((12,12))
```
